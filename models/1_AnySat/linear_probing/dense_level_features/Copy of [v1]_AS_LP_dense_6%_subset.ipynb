```python
import sys
import glob
import re
import math
import random
import shutil
import tarfile
import warnings
import tifffile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore")

def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

seed_everything(config.SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Computation backend configured: {device}")
```

```python
def setup_local_storage(config):
    splits = [
        ("train", config.DRIVE_TRAIN_FEAT_TAR, config.DRIVE_TRAIN_LAB_TAR),
        ("val",   config.DRIVE_VAL_FEAT_TAR,   config.DRIVE_VAL_LAB_TAR),
        ("test",  config.DRIVE_TEST_FEAT_TAR,  config.DRIVE_TEST_LAB_TAR)
    ]

    for split_name, drive_feat_tar, drive_lab_tar in splits:
        local_feat_dir = os.path.join(config.LOCAL_DATA_DIR, "features", split_name)
        local_lab_dir = os.path.join(config.LOCAL_DATA_DIR, "labels", split_name)

        os.makedirs(local_feat_dir, exist_ok=True)
        os.makedirs(local_lab_dir, exist_ok=True)

        # Skip if already extracted
        if len(os.listdir(local_feat_dir)) > 0 and len(os.listdir(local_lab_dir)) > 0:
            print(f"Split '{split_name}' already extracted locally. Skipping.")
            continue

        print(f"Extracting {split_name} data to local SSD...")

        # Process Features
        local_temp_tar = os.path.join(config.LOCAL_DATA_DIR, "temp.tar.gz")
        shutil.copy2(drive_feat_tar, local_temp_tar)
        with tarfile.open(local_temp_tar, "r:gz") as tar:
            tar.extractall(path=local_feat_dir)
        os.remove(local_temp_tar)

        # Process Labels
        shutil.copy2(drive_lab_tar, local_temp_tar)
        with tarfile.open(local_temp_tar, "r:gz") as tar:
            tar.extractall(path=local_lab_dir)
        os.remove(local_temp_tar)

    print("Local high-speed data setup complete.")

setup_local_storage(config)
```

```python
class AGBDataset(Dataset):
    def __init__(self, features_dir, labels_dir):
        # Recursive globbing handles arbitrary tar internal directory structures
        feature_files = glob.glob(os.path.join(features_dir, "**", "*.pt"), recursive=True)
        label_files = glob.glob(os.path.join(labels_dir, "**", "*.tif"), recursive=True)

        # Index labels by their 8-character ID
        label_dict = {}
        for lpath in label_files:
            match = re.search(r'([a-zA-Z0-9]{8})_agbm\.tif$', lpath)
            if match:
                label_dict[match.group(1)] = lpath

        self.samples = []
        for fpath in feature_files:
            match = re.search(r'_([a-zA-Z0-9]{8})_S1_10\.pt$', fpath)
            if match:
                fid = match.group(1)
                if fid in label_dict:
                    self.samples.append({"id": fid, "feature": fpath, "label": label_dict[fid]})

        print(f"Initialized {len(self.samples)} pairs in {os.path.basename(features_dir)}.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]

        # 1. Load the tensor (This will load the broken view, e.g., shape (1536,))
        feature = torch.load(sample["feature"], weights_only=True)

        # --- THE RECOVERY MAGIC ---
        if feature.shape != (256, 256, 1536):
            # Create an empty shell tensor with the correct float16 dtype
            recovered_tensor = torch.empty(0, dtype=feature.dtype)

            # Bind the shell directly to the raw 196MB storage block, forcing the correct shape
            recovered_tensor.set_(
                source=feature.untyped_storage(),
                storage_offset=0,
                size=(256, 256, 1536)
            )
            feature = recovered_tensor

        # Cast to float32 for training stability
        feature = feature.to(torch.float32)

        # 2. Load label: [256, 256] -> add channel -> [1, 256, 256]
        label = tifffile.imread(sample["label"])
        label = torch.from_numpy(label).unsqueeze(0).to(torch.float32)

        return feature, label, sample["id"]

# Initialize loaders
train_dataset = AGBDataset(os.path.join(config.LOCAL_DATA_DIR, "features/train"), os.path.join(config.LOCAL_DATA_DIR, "labels/train"))
val_dataset   = AGBDataset(os.path.join(config.LOCAL_DATA_DIR, "features/val"), os.path.join(config.LOCAL_DATA_DIR, "labels/val"))

train_loader = DataLoader(train_dataset, batch_size=config.BATCH_SIZE, shuffle=True, num_workers=config.NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_dataset, batch_size=config.BATCH_SIZE, shuffle=False, num_workers=config.NUM_WORKERS, pin_memory=True)
```

```python
class MetricTracker:
    def __init__(self):
        self.reset()

    def reset(self):
        self.n = 0
        self.sum_y = 0.0
        self.sum_y2 = 0.0
        self.sum_se = 0.0
        self.sum_ae = 0.0

    @torch.no_grad()
    def update(self, preds, targets):
        # Cast to float64 immediately for stability
        preds = preds.flatten().double()
        targets = targets.flatten().double()

        self.n += targets.numel()
        self.sum_y += targets.sum().item()
        self.sum_y2 += (targets ** 2).sum().item()

        errors = targets - preds
        self.sum_se += (errors ** 2).sum().item()
        self.sum_ae += errors.abs().sum().item()

    def compute(self):
        if self.n == 0: return {"MSE": 0.0, "RMSE": 0.0, "MAE": 0.0, "R2": 0.0}

        mse = self.sum_se / self.n
        rmse = math.sqrt(mse)
        mae = self.sum_ae / self.n

        # R2 Calculation
        ss_res = self.sum_se
        ss_tot = self.sum_y2 - (self.sum_y ** 2) / self.n
        r2 = 1.0 - (ss_res / ss_tot) if ss_tot > 0 else 0.0

        return {"MSE": mse, "RMSE": rmse, "MAE": mae, "R2": r2}
```

```python
class DenseLinearProbe(nn.Module):
    def __init__(self, in_features=1536):
        super().__init__()
        self.probe = nn.Linear(in_features, 1)

    def forward(self, x):
        # Input: [B, 256, 256, 1536]
        # Output: [B, 256, 256, 1]
        out = self.probe(x)
        # Permute back to standard image format [B, 1, 256, 256]
        return out.permute(0, 3, 1, 2)

model = DenseLinearProbe(in_features=config.IN_FEATURES).to(device)
criterion = nn.MSELoss()
optimizer = optim.AdamW(model.parameters(), lr=config.LEARNING_RATE, weight_decay=config.WEIGHT_DECAY)
```

```python
checkpoint_path = os.path.join(config.DRIVE_OUTPUT_DIR, "best_model.pt")
history_path = os.path.join(config.DRIVE_OUTPUT_DIR, "training_history.csv")
plot_path = os.path.join(config.DRIVE_OUTPUT_DIR, "training_curve.png")

start_epoch = 0
best_val_mse = float('inf')
patience_counter = 0
history = []

# --- Resume Checkpoint Logic ---
if os.path.exists(checkpoint_path):
    print("Found existing checkpoint. Resuming training...")
    chkpt = torch.load(checkpoint_path)
    model.load_state_dict(chkpt['model_state_dict'])
    optimizer.load_state_dict(chkpt['optimizer_state_dict'])
    start_epoch = chkpt['epoch'] + 1
    best_val_mse = chkpt['best_val_mse']
    patience_counter = chkpt.get('patience_counter', 0)

    if os.path.exists(history_path):
        history = pd.read_csv(history_path).to_dict('records')

# --- Training Loop ---
for epoch in range(start_epoch, config.EPOCHS):
    model.train()
    train_metrics = MetricTracker()

    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Train]")
    for features, labels, _ in pbar:
        features, labels = features.to(device), labels.to(device)

        optimizer.zero_grad()
        preds = model(features)
        loss = criterion(preds, labels)

        loss.backward()
        optimizer.step()

        train_metrics.update(preds, labels)
        pbar.set_postfix({"Loss": f"{loss.item():.4f}"})

    train_results = train_metrics.compute()

    # --- Validation ---
    model.eval()
    val_metrics = MetricTracker()
    with torch.no_grad():
        for features, labels, _ in tqdm(val_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Val]", leave=False):
            preds = model(features.to(device))
            val_metrics.update(preds, labels.to(device))

    val_results = val_metrics.compute()
    current_val_mse = val_results["MSE"]

    print(f"Epoch {epoch+1} | Train MSE: {train_results['MSE']:.4f} | Val MSE: {current_val_mse:.4f} | Val R2: {val_results['R2']:.4f}")

    # Record History
    epoch_data = {"Epoch": epoch+1, "Train_MSE": train_results['MSE'], **{f"Val_{k}": v for k, v in val_results.items()}}
    history.append(epoch_data)
    pd.DataFrame(history).to_csv(history_path, index=False)

    # Early Stopping & Checkpoint Save
    if current_val_mse < best_val_mse:
        best_val_mse = current_val_mse
        patience_counter = 0
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'best_val_mse': best_val_mse,
            'patience_counter': patience_counter,
            'val_metrics': val_results # Save metrics to retrieve during inference
        }, checkpoint_path)
    else:
        patience_counter += 1
        if patience_counter >= config.PATIENCE:
            print(f"Early stopping triggered at epoch {epoch+1}")
            break

# Plot Learning Curves
df = pd.DataFrame(history)
plt.figure(figsize=(10, 6))
plt.plot(df["Epoch"], df["Train_MSE"], label='Train MSE', marker='o')
plt.plot(df["Epoch"], df["Val_MSE"], label='Validation MSE', marker='s')
plt.title("Training vs Validation MSE")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.grid(True)
plt.savefig(plot_path)
plt.close()
print(f"Training complete. Artifacts saved to {config.DRIVE_OUTPUT_DIR}")
```

```python
# Setup test dataset and loader
test_dataset = AGBDataset(os.path.join(config.LOCAL_DATA_DIR, "features/test"), os.path.join(config.LOCAL_DATA_DIR, "labels/test"))
test_loader = DataLoader(test_dataset, batch_size=config.BATCH_SIZE, shuffle=False, num_workers=config.NUM_WORKERS)

# Load best isolated model
eval_model = DenseLinearProbe(in_features=config.IN_FEATURES).to(device)
chkpt = torch.load(checkpoint_path, map_location=device)
eval_model.load_state_dict(chkpt['model_state_dict'])
eval_model.eval()

# Run Evaluation
test_metrics = MetricTracker()
print("Running Inference on Test Set...")
with torch.no_grad():
    for features, labels, _ in tqdm(test_loader, desc="Testing"):
        preds = eval_model(features.to(device))
        test_metrics.update(preds, labels.to(device))

test_results = test_metrics.compute()

# Save final report to Drive
report_path = os.path.join(config.DRIVE_OUTPUT_DIR, "final_evaluation_report.txt")
with open(report_path, "w") as f:
    f.write(f"--- BEST MODEL PERFORMANCE REPORT ---\n")
    f.write(f"Best Checkpoint Epoch: {chkpt['epoch'] + 1}\n\n")
    f.write("VALIDATION METRICS:\n")
    for k, v in chkpt['val_metrics'].items(): f.write(f"  {k}: {v:.4f}\n")
    f.write("\nTEST METRICS (Unseen Data):\n")
    for k, v in test_results.items(): f.write(f"  {k}: {v:.4f}\n")

print(f"Test Evaluation Complete. R2 = {test_results['R2']:.4f}")
```

```python
def push_to_git(repo_dir, branch):
    try:
        pat = userdata.get('GITHUB_PAT')
        os.chdir(repo_dir)

        print("Staging changes...")
        subprocess.run(["git", "add", "."], check=True)

        # Check if there's anything to commit
        status = subprocess.run(["git", "status", "--porcelain"], capture_output=True, text=True)
        if not status.stdout.strip():
            print("No new changes to commit.")
            return

        print("Committing to repository...")
        subprocess.run(["git", "commit", "-m", "Auto-commit: Pipeline Training Complete"], check=True)

        print(f"Pushing to branch {branch}...")
        # Get remote URL and inject PAT for authenticated push
        remote_url = subprocess.run(["git", "config", "--get", "remote.origin.url"], capture_output=True, text=True).stdout.strip()
        auth_url = remote_url.replace("https://", f"https://{pat}@")

        subprocess.run(["git", "push", auth_url, branch], check=True)
        print("Git Push Successful!")

    except userdata.SecretNotFoundError:
        print("GITHUB_PAT missing. Commit bypassed.")
    except Exception as e:
        print(f"Git push failed: {e}")

push_to_git(GIT_REPO_DIR, BRANCH)
```
